

2. Encoding Categorical Variables in a Car Evaluation Dataset

```python
import pandas as pd

# Load the dataset with specified encoding
file_path = r'E:\data set\Car_sheet.csv'
car_data = pd.read_csv(file_path, encoding='ISO-8859-1')

car_data.head()
```

```python
# Identify categorical columns
categorical_columns = car_data.select_dtypes(include=['object']).columns
print("Categorical columns:", categorical_columns)
```

 Apply Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to each categorical column
car_data_label_encoded = car_data.copy()
for col in categorical_columns:
    car_data_label_encoded[col] = label_encoder.fit_transform(car_data[col])


car_data_label_encoded.head()
```

One-Hot Encoding

```python
# Apply one-hot encoding
car_data_hot_encoded = pd.get_dummies(car_data, columns=categorical_columns)
car_data_hot_encoded.head()
```

 Compare the Results

```python
# Compare the shape of the datasets
print("Original dataset shape:", car_data.shape)
print("Label encoded dataset shape:", car_data_label_encoded.shape)
print("One-hot encoded dataset shape:", car_data_hot_encoded.shape)
print("Label Encoded Dataset:")
car_data_label_encoded.head()

print("One-Hot Encoded Dataset:")
car_data_hot_encoded.head()
```
