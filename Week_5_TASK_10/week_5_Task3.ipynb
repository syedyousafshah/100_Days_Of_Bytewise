3. Scaling Features in the Wine Quality Dataset

```python
import pandas as pd
file_path = r'E:\data set\DATA SET\wine-quality.csv'  
wine_data = pd.read_csv(file_path)

wine_data.head()
```

```python
# Separate features and target variable
X = wine_data.drop(columns=['quality']) 
y = wine_data['quality']

X.head()
```

Apply Normalization

```python
from sklearn.preprocessing import MinMaxScaler
# Initialize the scaler
min_max_scaler = MinMaxScaler()
# Apply normalization
X_normalized = min_max_scaler.fit_transform(X)
# Convert the result back to a DataFrame
X_normalized = pd.DataFrame(X_normalized, columns=X.columns)
X_normalized.head()
```

Apply Standardization

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
standard_scaler = StandardScaler()
# Apply standardization
X_standardized = standard_scaler.fit_transform(X)
# Convert the result back to a DataFrame
X_standardized = pd.DataFrame(X_standardized, columns=X.columns)
X_standardized.head()
```

Analyze the Effect of Scaling

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for the plots
sns.set(style="whitegrid")

# Plot the distribution of the original, normalized, and standardized features
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(12, 18))

# Original data distribution
sns.boxplot(data=X, ax=axes[0], palette="Set3")
axes[0].set_title('Original Data Distribution', fontsize=16)
axes[0].tick_params(axis='x', rotation=45)

# Normalized data distribution
sns.boxplot(data=X_normalized, ax=axes[1], palette="Set2")
axes[1].set_title('Normalized Data Distribution', fontsize=16)
axes[1].tick_params(axis='x', rotation=45)

# Standardized data distribution
sns.boxplot(data=X_standardized, ax=axes[2], palette="Set1")
axes[2].set_title('Standardized Data Distribution', fontsize=16)
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
