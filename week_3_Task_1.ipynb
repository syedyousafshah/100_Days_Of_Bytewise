```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
file_path = r'E:\data set\tesla-stock-price.csv'
HPrices = pd.read_csv(file_path)

# Print column names to verify existence of 'date' column
print("Columns in HPrices DataFrame:", HPrices.columns)

# Example: Sorting and plotting based on 'date'
# Assuming 'date' column is already in datetime format
if 'date' in HPrices.columns:
    HPrices.sort_values(by='date', inplace=True)  # Ensure column name matches exactly
    plt.figure(figsize=(12, 6))
    plt.plot(HPrices['date'], HPrices['close'], label='Closing Price')
    plt.title('Tesla Stock Closing Prices')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()
else:
    print("The 'date' column does not exist in HPrices DataFrame. Check your data.")
```

```python
import pandas as pd
grouped_dataset = titanic.groupby(['Pclass', 'Survived']).size().unstack()
# Display the grouped dataset
print(grouped_dataset)
```

```python
# Assuming grouped_dataset is already defined
x_axis = grouped_dataset.index
survived = grouped_dataset[0]  #  0 represents 'survived'
not_survived = grouped_dataset[1]  
print("Survived:")
print(survived)
print("\nNot Survived:")
print(not_survived)
```

```python
import matplotlib.pyplot as plt

# Assuming x_axis, survived, and not_survived are defined as in the previous steps

fig, (plt1, plt2) = plt.subplots(1, 2, figsize=(12, 6))  # Adjusted figsize for better visibility

plt1.set_title('Survived Vs Passenger Class')
plt1.bar(x_axis, survived, edgecolor='black', color='green')
plt1.set_xlabel('Passenger Class')
plt1.set_ylabel('People Survived')
plt1.set_ylim(0, max(survived) + 50)  # Adjusted ylim to include all data points

plt2.set_title('Not Survived Vs Passenger Class')
plt2.bar(x_axis, not_survived, edgecolor='black', color='red')  # Changed color for not survived
plt2.set_xlabel('Passenger Class')
plt2.set_ylabel('People Not Survived')
plt2.set_ylim(0, max(not_survived) + 50)  # Adjusted ylim to include all data points

plt.tight_layout()  # Ensures labels don't overlap
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t_axis = np.linspace(0, 5, 50)
sin_function = np.sin(np.pi * t_axis)
cos_function = np.cos(np.pi * t_axis)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))  
# Plot in the first subplot
ax1.plot(t_axis, sin_function, 'r-', label='sin function')
ax1.set_xlabel("Time axis")
ax1.set_ylabel("Output function")
ax1.set_title("Subplot 1")
ax1.set_xlim(-0.5, 5.5) 
ax1.set_ylim(-1.2, 1.2)  
ax1.grid(True)
ax1.legend(loc='upper right')  

# Plot in the second subplot
ax2.plot(t_axis, cos_function, 'g-', label='cos function')
ax2.set_xlabel("Time axis")
ax2.set_ylabel("Output function")
ax2.set_title("Subplot 2")
ax2.set_xlim(-0.5, 5.5)  
ax2.set_ylim(-1.2, 1.2)  
ax2.grid(True)
ax2.legend(loc='upper right')  # Adjusted legend location

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset from Seaborn
tips = sns.load_dataset('tips')

# Set up the figure and axes
plt.figure(figsize=(10, 6))
plt.title('Distribution of Total Bill with KDE Overlay')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')

# Create histogram with KDE overlay using Seaborn
sns.histplot(tips['total_bill'], kde=True, color='skyblue', edgecolor='black')

# Customize legend and grid
plt.legend(['KDE'])
plt.grid(True)

plt.show()
```
